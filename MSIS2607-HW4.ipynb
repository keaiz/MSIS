### Question 3

For each Job, return the rows of those people who, within their group, know the most Languages. (In Camino, pick the correct total number of people who satisfy this condition)

```python
df.groupby('Job')['Languages'].max()
```

Above, we have the maximum number of values each group possesses. For example, within the part-time category (0.5); the highest number of languages known among anyone in that category is 6. 

To make this work for any program with any data, let's store the maximum value within each category in a variable.

```python
no_job = df.groupby('Job')['Languages'].max()[0]   #stores score of 7
```

```python
part_time = df.groupby('Job')['Languages'].max()[0.5]   #stores score of 6
```

```python
full_time = df.groupby('Job')['Languages'].max()[1]    #stores score of 7
```

Now let's make a boolean to yield the desired results, using our new variables.

```python
df[((df.Languages == no_job)&(df.Job==0.0))|
   ((df.Languages == part_time)&(df.Job==0.5))|((df.Languages == full_time)&(df.Job==1.0))]
```

These are the rows of the students who know the highest number of languages within their respective groups (The groups being based on their work status). There are five students total who satisfy this requirement.
