##### Section 1.1: Trump-related Tweets

Our first step will be to read the txt files into a python string. We have a three, each containing a list of positive, negative and stop words. We also have one more titled Trump.txt containing tweets containing the keyword 'Trump' which we extracted in real-time using a twitter development account. 

```python
pos = open('positive.txt','r')
poscontent = pos.read()  #text of all positive words (not in list form yet due to messiness)
pos.close() 

neg = open('negative.txt','r')
negcontent = neg.read()  #text of all negative words (not in list form yet due to messiness)
neg.close() 

stop = open('stopwords.txt','r')
stopcontent = stop.read()  #text of all stop words (not in list form yet due to messiness)
stop.close() 

trump = open('Trump.txt','r')
trump_content = trump.read()  #text of all trump-related tweets (not in list form yet due to messiness)
trump.close() 
```

Let's split the positive, negative, and stopwords into a list. It's easy to do since they're delimited only by a newline character. 

```python
poslist=poscontent.split('\n')
neglist=negcontent.split('\n')
stoplist=stopcontent.split('\n')  #Here we're making a list delimited at the newlines characters/ removing newline
```

```python
trump_content   # raw text of all trump related tweets (not yet in list form)
```

Now let's use some string methods to clean up the messy list of tweets displayed above. We will clean as much as possible. 

It's important to make the entire string lowercase, as well as to remove any unnecessary puncuation. This is important, because when counting the instances of positive, negative, and stopwords; any puncuation or case difference could confuse our counter. 

<b>For example: *Good* $\neq$ *good* $\neq$ *good!*</b>

```python
#Let's use some string methods to turn all words 
#into lower case and remove puncuation so as to remove all unnessecary noise

trump_content=trump_content.lower()


badchars='.,!?"///#[]()*$:'
for i in badchars:
    trump_content=trump_content.replace(i,'')
```

The string of tweets should now be turned into a list of words. This way we can iterate through all the words and count how many positives and negatives appear. 

```python
trumplist=trump_content.split()
trumplist  # here we turned the trump text into a list of words. 
```

In the preview above, we still see lots of unnecessary noise. Let's use some further string methods to remove any web links, usernames, retweets, and what not. 

```python
import random
random.seed(1)   #set seed for consistent number generation
for i in trumplist:
    if i.startswith("https") or i.endswith("b'rt") or i.startswith("@") or "\\" in i or 'xe2' in i:
        trumplist.remove(i)
        
#Use the startswith() and endswith() methods to remove noise. Since user names begin with '@' 
# and links begin with 'http', etc


for i in trumplist:
    if i.startswith("https"):
        trumplist.remove(i)
        
    if i.startswith("@"):
        trumplist.remove(i)
        

#Loop one more time to make sure everything is accounted for 
```

```python

for i in trumplist:
    print(i)  #Most of the noise has been removed
```

After breaking up the tweets into a list of individual words, we observe above that most of the noise has been eliminated using our string methods. As a consequence of using a list so large, there will still be a little bit of noise remaining that couldn't be accounted for. This is okay, and is to be expected. We'll consider that later in our analysis. But for now, just from eyeballing the list, it seems we cleaned up most of the document. 

```python
len(trumplist)  #The total number of words in the tweet document
```

Now we can begin using for loops and accumulators to iterate through the list of words in trumplist; and then count how many positive, negative, and stop words appear in Trump-related tweets. We'll create a fourth category called 'other words' to account for all words that are neither positive, negative, nor stop words. 

```python
poswords=0
for word in trumplist:
    if word in poslist:
        poswords=poswords+1
        
negwords=0
for word in trumplist:
    if word in neglist:
        negwords=negwords+1
        
stopwords=0
for word in trumplist:
    if word in stoplist:
        stopwords=stopwords+1
        
otherwords=0
for word in trumplist:
    if word not in poslist and word not in neglist and word not in stoplist:
        otherwords=otherwords+1
```

```python
otherwords
```

Just from eyeballing our initial results, it appears that *Positive Words* $\gt$ *Negative Words*.
This is an interesting observation because it seems like 'Trump', being a politically charged keyword, would yield a more negative response rather than a positive one. 

Upon observing the list of positive keywords in positive.txt; we can see that the word *trump* (Not the individual) is listed as a positive word. This is a source of confusion for our analysis.
Let's count *positive words* one more time, now counting all positive words **Execept For** Trump.

```python
poswords_updated=0
for word in trumplist:
    if word in poslist and word != 'trump':
        poswords_updated=poswords_updated+1
```

```python
poswords_updated
```

Our number of positive words was cut in half. This yields a more reasonable result and ratio. We'll analyze our results at the end of our report.
